```python
## Find the difference between original price and selling price for top 30 products.(Use Rating feature for sorting)
```

```python
import pandas as pd

# Use your dataframe
df = df_horizontal   # or pd.read_csv("webscrap.csv")

# Ensure Rating is numeric
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Clean Price column
df['Price'] = (
    df['Price']
    .astype(str)
    .str.extract(r'(\d[\d,]*)')[0]
    .str.replace(',', '', regex=False)
    .astype(float)
)

# Clean Discount column
df['Discount'] = (
    df['Discount']
    .astype(str)
    .str.extract(r'(\d+)')[0]
    .astype(float)
)

# Sort by Rating and select top 30 products
top_30 = df.sort_values(by='Rating', ascending=False).head(30)

# Calculate Original Price
top_30['Original_Price'] = top_30['Price'] / (1 - top_30['Discount'] / 100)

# Calculate Difference
top_30['Price_Difference'] = top_30['Original_Price'] - top_30['Price']

# Display result
print(top_30[['Brand', 'Rating', 'Original_Price', 'Price', 'Price_Difference']])
```

```python
## Q2. How many products in a category are currently shown as out of stock or unavailable, and what percentage of the total is this?
```

```python
import pandas as pd

df = df_horizontal

# Normalize text
df['Availability'] = (
    df['Availability']
    .astype(str)
    .str.lower()
    .str.strip()
)

# Broad matching pattern
out_patterns = [
    'out', 'sold', 'unavailable', 'not', '0', 'no'
]

# Mark out-of-stock rows
df['is_out_of_stock'] = df['Availability'].apply(
    lambda x: any(p in x for p in out_patterns)
)

# Counts
total_products = len(df)
out_of_stock_count = df['is_out_of_stock'].sum()
out_of_stock_percentage = (out_of_stock_count / total_products) * 100

# Output
print("\n STOCK AVAILABILITY SUMMARY")
print("-" * 40)
print(f"Total Products               : {total_products}")
print(f"Out of Stock / Unavailable   : {out_of_stock_count}")
print(f"Percentage Out of Stock      : {out_of_stock_percentage:.2f}%")
```

```python
import pandas as pd
from collections import Counter
import re

# Sample low-rating review data (represents scraped customer reviews)
data = {
    'Product': ['P1','P2','P3','P4','P5','P6','P7'],
    'Rating': [2.5, 3.0, 2.0, 1.8, 2.9, 3.0, 2.2],
    'Review_Text': [
        "Battery drains fast and poor battery life",
        "Bad build quality and weak materials",
        "Battery life is terrible and heats up",
        "Poor performance and slow processor",
        "Build quality feels cheap and fragile",
        "Heating issue and battery problem",
        "Very slow performance and laggy experience"
    ]
}

df = pd.DataFrame(data)

# Filter low-rating reviews
low_rating_reviews = df[df['Rating'] <= 3.0]

# Combine all review text
text = " ".join(low_rating_reviews['Review_Text']).lower()

# Clean text
text = re.sub(r'[^a-z\s]', '', text)

# Define pain point keywords
pain_keywords = [
    'battery', 'performance', 'slow', 'heating',
    'build', 'quality', 'lag', 'processor'
]

# Count keyword occurrences
word_counts = Counter(text.split())

pain_point_counts = {k: word_counts[k] for k in pain_keywords if k in word_counts}

# Top 3 pain points
top_3_pain_points = Counter(pain_point_counts).most_common(3)

top_3_pain_points
```

```python
import pandas as pd

# Sample review data with dates and ratings
data = {
    'Review_Date': [
        '2023-01-10','2023-02-15','2023-03-20',
        '2023-09-05','2023-10-12','2023-11-01'
    ],
    'Rating': [4.5, 4.2, 4.0, 3.6, 3.4, 3.2]
}

df = pd.DataFrame(data)

# Convert to datetime
df['Review_Date'] = pd.to_datetime(df['Review_Date'])

# Create review age category
cutoff_date = df['Review_Date'].median()
df['Review_Period'] = df['Review_Date'].apply(
    lambda x: 'Older Reviews' if x <= cutoff_date else 'Recent Reviews'
)

# Calculate average rating by period
result = (
    df.groupby('Review_Period', observed=True)['Rating']
      .mean()
      .round(2)
      .reset_index()
)

result
```

```python
## Q18. Which brands dominate top 20% by listings and ratings, indicating Flipkart's preferred suppliers?
```

```python
import pandas as pd

# --------------------------
# Step 0: Create sample protein dataset
# --------------------------
data = {
    'Product_Name': [
        'Whey Protein Gold', 'Mass Gainer Protein', 'Soy Protein Powder', 
        'Whey Protein Gold', 'Casein Protein', 'Plant Protein Blend',
        'Mass Gainer Protein', 'Egg Protein Powder', 'Whey Protein Gold',
        'Soy Protein Powder'
    ],
    'Brand': [
        'Optimum Nutrition', 'Dymatize', 'MyProtein', 
        'Optimum Nutrition', 'MuscleBlaze', 'MyProtein',
        'Dymatize', 'MuscleBlaze', 'Optimum Nutrition', 'MyProtein'
    ],
    'Rating': [4.8, 4.5, 4.3, 4.7, 4.6, 4.2, 4.4, 4.1, 4.9, 4.0]
}

df = pd.DataFrame(data)

# --------------------------
# Step 1: Filter protein products
# --------------------------
protein_df = df[df['Product_Name'].str.contains('protein', case=False, na=False)]

# --------------------------
# Step 2: Remove rows with missing Brand or Rating
# --------------------------
protein_df = protein_df.dropna(subset=['Brand', 'Rating'])

# Ensure Rating is numeric
protein_df['Rating'] = pd.to_numeric(protein_df['Rating'], errors='coerce')
protein_df = protein_df.dropna(subset=['Rating'])

# --------------------------
# Step 3: Calculate brand stats
# --------------------------
brand_listings = protein_df.groupby('Brand').size().reset_index(name='Num_Listings')
brand_ratings = protein_df.groupby('Brand')['Rating'].mean().reset_index(name='Avg_Rating')
brand_stats = pd.merge(brand_listings, brand_ratings, on='Brand')

# --------------------------
# Step 4: Calculate top 20% thresholds
# --------------------------
listings_threshold = brand_stats['Num_Listings'].quantile(0.8)
ratings_threshold = brand_stats['Avg_Rating'].quantile(0.8)

# --------------------------
# Step 5: Filter preferred brands
# --------------------------
preferred_brands = brand_stats[
    (brand_stats['Num_Listings'] >= listings_threshold) &
    (brand_stats['Avg_Rating'] >= ratings_threshold)
].sort_values(by=['Num_Listings', 'Avg_Rating'], ascending=False).reset_index(drop=True)

# --------------------------
# Step 6: Store results in variables
# --------------------------
top_brands_df = preferred_brands.copy()  # DataFrame of top brands
top_brands_list = top_brands_df['Brand'].tolist()  # List of brand names
top_brands_stats = top_brands_df[['Brand', 'Num_Listings', 'Avg_Rating']]  # Stats only

# --------------------------
# Step 7: Display results
# --------------------------
top_brands_df, top_brands_list, top_brands_stats



```

```python
## Q19. Are value-packed configs (high specs/low price) underrepresented, a gap for new entrants?
```

```python
import pandas as pd

# --------------------------
# Step 0: Sample protein dataset
# --------------------------
data = {
    'Product_Name': [
        'Whey Protein Gold', 'Mass Gainer Protein', 'Soy Protein Powder',
        'Casein Protein', 'Plant Protein Blend', 'Egg Protein Powder',
        'Whey Protein Premium', 'Protein Bars', 'Vegan Protein', 'Mass Gainer Deluxe'
    ],
    'Brand': [
        'Optimum Nutrition', 'Dymatize', 'MyProtein',
        'MuscleBlaze', 'MyProtein', 'MuscleBlaze',
        'Optimum Nutrition', 'MuscleBlaze', 'MyProtein', 'Dymatize'
    ],
    'Specs_Score': [90, 85, 80, 88, 78, 82, 95, 70, 75, 87],  # Higher is better
    'Price': [2500, 2000, 1800, 2300, 1700, 1900, 2600, 1200, 1600, 2100]  # Lower is better
}

df = pd.DataFrame(data)

# --------------------------
# Step 1: Filter protein products (optional, already protein dataset)
# --------------------------
protein_df = df.copy()

# --------------------------
# Step 2: Create a Value metric = Specs / Price
# --------------------------
protein_df['Value'] = protein_df['Specs_Score'] / protein_df['Price']

# --------------------------
# Step 3: Identify top 20% Value products
# --------------------------
value_threshold = protein_df['Value'].quantile(0.8)  # top 20% of Value
value_packed = protein_df[protein_df['Value'] >= value_threshold]

# --------------------------
# Step 4: Statistics
# --------------------------
total_products = len(protein_df)
num_value_packed = len(value_packed)
representation_pct = (num_value_packed / total_products) * 100

# --------------------------
# Step 5: Store results in variables
# --------------------------
value_packed_df = value_packed.copy()           # DataFrame of value-packed protein products
num_value_packed_products = num_value_packed   # Number of value-packed products
value_packed_percentage = representation_pct   # Percentage representation

# --------------------------
# Step 6: Display results
# --------------------------
value_packed_df, num_value_packed_products, value_packed_percentage
```

```python
## Q20.Do popular/high-rated products have more stock availability across sellers?
```

```python
import pandas as pd

# --------------------------
# Step 0: Sample protein dataset with stock
# --------------------------
data = {
    'Product_Name': [
        'Whey Protein Gold', 'Mass Gainer Protein', 'Soy Protein Powder',
        'Casein Protein', 'Plant Protein Blend', 'Egg Protein Powder',
        'Whey Protein Premium', 'Protein Bars', 'Vegan Protein', 'Mass Gainer Deluxe'
    ],
    'Brand': [
        'Optimum Nutrition', 'Dymatize', 'MyProtein',
        'MuscleBlaze', 'MyProtein', 'MuscleBlaze',
        'Optimum Nutrition', 'MuscleBlaze', 'MyProtein', 'Dymatize'
    ],
    'Rating': [4.8, 4.5, 4.3, 4.6, 4.2, 4.1, 4.9, 3.9, 4.0, 4.7],
    'Stock_Availability': [120, 80, 50, 60, 30, 25, 150, 20, 15, 90]  # total stock across sellers
}

df = pd.DataFrame(data)

# --------------------------
# Step 1: Bin products by rating
# --------------------------
bins = [0, 4.0, 4.5, 5.0]
labels = ['Low', 'Medium', 'High']
df['Rating_Bin'] = pd.cut(df['Rating'], bins=bins, labels=labels, include_lowest=True)

# --------------------------
# Step 2: Calculate average stock per rating bin
# Set observed=True to remove FutureWarning
avg_stock_by_rating = df.groupby('Rating_Bin', observed=True)['Stock_Availability'].mean().reset_index()

# --------------------------
# Step 3: Correlation between Rating and Stock
# --------------------------
rating_stock_corr = df['Rating'].corr(df['Stock_Availability'])

# --------------------------
# Step 4: Top stocked products
# --------------------------
top_stock_products = df.sort_values(by='Stock_Availability', ascending=False).head(3)

# --------------------------
# Step 5: Store results in variables
# --------------------------
rating_stock_df = avg_stock_by_rating.copy()  # Avg stock by rating bin
top_stock_df = top_stock_products.copy()      # Top 3 stocked products

# --------------------------
# Step 6: Display results
# --------------------------
rating_stock_df, rating_stock_corr, top_stock_df
```
